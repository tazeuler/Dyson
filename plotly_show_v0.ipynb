```python
import pandas as pd
import numpy as np
import random
import string
import seaborn as sns
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Generate synthetic data for 100 bonds
n_bonds = 100
isin_numbers = [''.join(random.choices(string.ascii_uppercase + string.digits, k=12)) for _ in range(n_bonds)]
sectors = np.random.choice(['Finance', 'Technology', 'Healthcare', 'Energy', 'Consumer Goods', 'Telecommunications'], n_bonds)
spread_levels = np.random.uniform(1.0, 3.0, n_bonds)

# Create a DataFrame with the synthetic data
df = pd.DataFrame({
    'ISIN': isin_numbers,
    'Sector': sectors,
    'Spread Level': spread_levels
})

# Calculate mean of spread levels by sector
mean_df = df.groupby('Sector')['Spread Level'].mean().reset_index()

# Create a pairplot with spread levels and mean by sector
g = sns.pairplot(df, x_vars='Spread Level', y_vars='Sector', hue='Sector', height=5, aspect=1.5)
g.fig.suptitle('Spread Levels and Mean by Sector', y=1.02)

plt.show()
```
